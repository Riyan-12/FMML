```python
# String Methods
string = "Hello, World!"

# len(): returns the length of the string
print(len(string))

# upper(): converts the string to uppercase
print(string.upper())

# lower(): converts the string to lowercase
print(string.lower())

# split(): splits the string into a list of substrings
print(string.split(","))

# replace(): replaces a substring with another substring
print(string.replace("Hello", "Hi"))
```

```python
# List Methods
my_list = [1, 2, 3, 4, 5]

# append(): adds an element to the end of the list
my_list.append(6)
print(my_list)

# extend(): extends the list by appending elements from another iterable
my_list.extend([7, 8, 9])
print(my_list)

# insert(): inserts an element at a specified index
my_list.insert(0, 0)
print(my_list)

# remove(): removes the first occurrence of an element
my_list.remove(5)
print(my_list)

# reverse(): reverses the order of elements in the list
my_list.reverse()
print(my_list)
```

```python
# Tuple Methods
my_tuple = (1, 2, 3, 4, 5)

# index(): returns the index of the first occurrence of a value
print(my_tuple.index(3))

# count(): returns the number of occurrences of a value
print(my_tuple.count(4))

```

```python
# Dictionary Methods
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}

# keys(): returns a list of all the keys in the dictionary
print(my_dict.keys())

# values(): returns a list of all the values in the dictionary
print(my_dict.values())

# items(): returns a list of key-value pairs as tuples
print(my_dict.items())

# get(): returns the value associated with a specified key
print(my_dict.get('name'))

# pop(): removes and returns the value associated with a specified key
print(my_dict.pop('age'))
```

```python
# Set Methods
my_set = {1, 2, 3, 4, 5}

# add(): adds an element to the set
my_set.add(6)
print(my_set)

# remove(): removes an element from the set
my_set.remove(3)
print(my_set)

# union(): returns a new set with all elements from both sets
other_set = {4, 5, 6, 7, 8}
print(my_set.union(other_set))

# intersection(): returns a new set with common elements from both sets
print(my_set.intersection(other_set))

# difference(): returns a new set with elements in the first set but not in the second set
print(my_set.difference(other_set))
```

```python
def reverse_and_concatenate(A, B):
    reversed_A = A[::-1]  # Reverses the first string A
    result = reversed_A + B  # Concatenates the reversed A with the second string B
    return result

# Example usage:
string_A = "Hello"
string_B = "World"
final_string = reverse_and_concatenate(string_A, string_B)
print(final_string)
```

```python
def calculate_payment(workers):
    max_hours = 0
    max_hours_name = ""

    for worker in workers:
        if worker[2].lower() == "female" and worker[1] > max_hours:
            max_hours = worker[1]
            max_hours_name = worker[0]

    if max_hours_name != "":
        payment = max_hours * 20
        print("The female worker who worked the most hours is:", max_hours_name)
        print("She should be paid:", payment, "$")
    else:
        print("No female workers found in the list.")

# Example usage:
worker_list = [
    ["Alice", 40, "female"],
    ["Bob", 35, "male"],
    ["Catherine", 50, "female"],
    ["David", 45, "male"]
]

calculate_payment(worker_list)
```

```python
import numpy as np

row = np.array([[1, 1], [2, 2]])
matrix = np.vstack(np.repeat(row, 5, axis=0))

# Creating another matrix for dot product
other_matrix = np.random.randint(1, 10, size=(2, 5))

# Finding the dot product
dot_product = np.dot(matrix, other_matrix)

print("Matrix shape:", matrix.shape)
print("Other matrix shape:", other_matrix.shape)
print("Dot product shape:", dot_product.shape)
print("Dot product matrix:")
print(dot_product)
```
